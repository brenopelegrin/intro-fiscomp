```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.io import FortranFile
import numpy as np
```

```python
plt.style.use("_classic_test_patch")
```

```python
def readf_xy(filePath: str):
    data = []
    with open(filePath, 'r') as f:
        for line in f:
            data.append(line.rstrip().split())
    array = np.array(data, dtype=float)
    return array

data1 = readf_xy('saida-d-13687303.dat')
```

```python
f1, ax1 = plt.subplots()

ax1.plot(data1[:, 0], data1[:, 1], linewidth=0.5)
ax1.set_ylabel('Entropia ($S$)')
ax1.set_xlabel('NÃºmero de passos ($N$)')
f1.savefig('./tarefa-d-graf-1-13687303.pdf', format='pdf', dpi=300, bbox_inches='tight')
```
