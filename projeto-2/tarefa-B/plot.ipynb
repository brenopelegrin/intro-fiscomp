```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.io import FortranFile
import numpy as np
```

```python
plt.style.use("_classic_test_patch")
```

```python
def readf_xy(filePath: str):
    data = []
    with open(filePath, 'r') as f:
        for line in f:
            data.append(line.rstrip().split())
    array = np.array(data, dtype=float)
    return array[array[:,1] != 0]

data1 = readf_xy('saida-b-13687303-hist-p1over2.dat')
data2 = readf_xy('saida-b-13687303-hist-p1over3.dat')
data3 = readf_xy('saida-b-13687303-hist-p1over4.dat')
data4 = readf_xy('saida-b-13687303-hist-p1over5.dat')
```

```python
f1, ax1 = plt.subplots()
ax1.plot(data1[:, 0], data1[:, 1], label='$p=1/2$', alpha=1)
ax1.plot(data2[:, 0], data2[:, 1], label='$p=1/3$', alpha=1)
ax1.plot(data3[:, 0], data3[:, 1], label='$p=1/4$', alpha=1)
ax1.plot(data4[:, 0], data4[:, 1], label='$p=1/5$', alpha=1)
ax1.grid()
ax1.set_ylabel('Quantidade de andarilhos')
ax1.set_xlabel('Posio final x')
ax1.legend(loc='upper center', bbox_to_anchor=(0.5, -0.13),
          fancybox=True, shadow=True, ncol=5)
f1.savefig('./tarefa-b-graf-1-13687303.pdf', format='pdf', dpi=300, bbox_inches='tight')
```
