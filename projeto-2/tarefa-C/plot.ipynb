```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.io import FortranFile
import numpy as np
```

```python
plt.style.use("_classic_test_patch")
```

```python
def readf_xy(filePath: str):
    data = []
    with open(filePath, 'r') as f:
        for line in f:
            data.append(line.rstrip().split())
    array = np.array(data, dtype=float)
    return array

data1 = readf_xy('saida-c-13687303-n1e1.dat')
data2 = readf_xy('saida-c-13687303-n1e2.dat')
data3 = readf_xy('saida-c-13687303-n1e3.dat')
data4 = readf_xy('saida-c-13687303-n1e4.dat')
data5 = readf_xy('saida-c-13687303-n1e5.dat')
data6 = readf_xy('saida-c-13687303-n1e6.dat')
```

```python
f1, ax1 = plt.subplots()

ax1.scatter(data6[:, 0], data6[:, 1], label='$N=10^6$', alpha=0.3, s=10, color='black')
ax1.scatter(data5[:, 0], data5[:, 1], label='$N=10^5$', alpha=0.3, s=10, color='yellow')
ax1.scatter(data4[:, 0], data4[:, 1], label='$N=10^4$', alpha=0.3, s=10, color='green')
ax1.scatter(data3[:, 0], data3[:, 1], label='$N=10^3$', alpha=0.3, s=10, color='red')
ax1.scatter(data2[:, 0], data2[:, 1], label='$N=10^2$', alpha=0.3, s=10, color='blue')
ax1.scatter(data1[:, 0], data1[:, 1], label='$N=10^1$', alpha=0.3, s=10, color='cyan')
ax1.set_ylabel('Posição y')
ax1.set_xlabel('Posição x')
ax1.legend()
f1.savefig('./tarefa-c-graf-1-13687303.pdf', format='pdf', dpi=300, bbox_inches='tight')
```
