```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.io import FortranFile
import numpy as np
```

```python
plt.style.use("_classic_test_patch")
```

```python
def readf_xy(filePath: str):
    data = []
    with open(filePath, 'r') as f:
        for line in f:
            data.append(line.rstrip().split())
    array = np.array(data, dtype=float)
    return array

data1 = readf_xy('saida-c-13687303-pos-m1000-n10.dat')
data2 = readf_xy('saida-c-13687303-pos-m1000-n100.dat')
data3 = readf_xy('saida-c-13687303-pos-m1000-n1000.dat')
data4 = readf_xy('saida-c-13687303-pos-m1000-n100000.dat')
data5 = readf_xy('saida-c-13687303-pos-m1000-n1000000.dat')
data6 = readf_xy('saida-c-13687303-pos-m1000-n10000000.dat')
```

```python
f1, ax1 = plt.subplots()
ax1.scatter(data1[:, 0], data1[:, 1], label='$p=1/2$', alpha=1)
ax1.scatter(data2[:, 0], data2[:, 1], label='$p=1/3$', alpha=1)
ax1.scatter(data3[:, 0], data3[:, 1], label='$p=1/4$', alpha=1)
ax1.scatter(data4[:, 0], data4[:, 1], label='$p=1/5$', alpha=1)
ax1.grid()
ax1.legend(loc='upper center', bbox_to_anchor=(0.5, -0.07),
          fancybox=True, shadow=True, ncol=5)
```
