```python
from sympy import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x,f,y = symbols('x f y')
```

```python
f_a = exp(x/2)*tan(2*x)
d1f_a = diff(f_a, x)
d2f_a = diff(d1f_a, x)
d3f_a = diff(d2f_a, x)
```

```python
d1f_a
```

```python
d2f_a
```

```python
d3f_a
```

```python
d1f_a_val = str(N(d1f_a.subs(x, 0.5), 12))
d2f_a_val = str(N(d2f_a.subs(x, 0.5), 13))
d3f_a_val = str(N(d3f_a.subs(x, 0.5), 14))
print(f"d1:   {d1f_a_val}")
print(f"d2:  {d2f_a_val}")
print(f"d3: {d3f_a_val}")
```
