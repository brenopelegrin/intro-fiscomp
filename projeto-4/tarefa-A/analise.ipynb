```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Importando os dados em CSV do Fortran

```python
theta_e = pd.read_csv('./theta_t.csv', sep=',', header=0)
e_e = pd.read_csv('./e_t.csv', sep=',', header=0)
theta_ec = pd.read_csv('./theta_ec_t.csv', sep=',', header=0)
e_ec = pd.read_csv('./e_ec_t.csv', sep=',', header=0)
```

# Gráficos

## Método de Euler

- Gráfico de $\theta(t)$

```python
f1, ax1 = plt.subplots()
ax1.plot(theta_e['t'], theta_e['theta'])
ax1.set_xlabel('Tempo (s)')
ax1.set_ylabel('$\\theta$ (rad)')
ax1.grid()
ax1.legend()
```

- Gráfico de $E(t)$

```python
f2, ax2 = plt.subplots()
ax2.plot(e_e['t'], e_e['e'], label='$E(t)$')
ax2.set_xlabel('Tempo (s)')
ax2.axhline(np.mean(e_e['e']), color='red', label='E média')
ax2.axhline(e_e['e'][0], color='green', label='$E_0$')
ax2.set_ylabel('Energia (J)')
ax2.grid()
ax2.legend(loc='upper center', bbox_to_anchor=(0.5, -0.12),
          fancybox=True, shadow=True, ncol=3)
```

## Método de Euler-Cromer

- Gráfico de $\theta(t)$

```python
f3, ax3 = plt.subplots()
ax3.plot(theta_ec['t'], theta_ec['theta_ec'])
ax3.set_ylabel("$\\theta$ (rad)")
ax3.set_xlabel('Tempo (s)')
ax3.grid()
ax3.legend()
```

- Gráfico de $E(t)$

```python
f4, ax4 = plt.subplots()
ax4.plot(e_ec['t'], e_ec['e_ec'], label='$E (t)$')
ax4.axhline(np.mean(e_ec['e_ec']), color='red', label='E média')
ax4.axhline(e_ec['e_ec'][0], color='green', label='$E_0$')
ax4.set_ylabel('Energia (J)')
ax4.set_xlabel('Tempo (s)')
ax4.grid()
ax4.legend(loc='upper center', bbox_to_anchor=(0.5, -0.12),
          fancybox=True, shadow=True, ncol=3)
```
